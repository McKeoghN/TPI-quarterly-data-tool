```python
OPH = pd.read_csv("../out/OPH_Processed.csv")

print(OPH)
```

```python
fig = px.line(OPH, 
              x="Year", 
              y="GDP per Hour Worked", 
              color="Country",
              markers=True,
              title="GDP per Hour Worked (2015=100) Over Time")

fig.add_shape(
    go.layout.Shape(
        type="line",
        x0=2015, x1=2015,  # Vertical line at 2015
        y0=OPH["GDP per Hour Worked"].min(), 
        y1=OPH["GDP per Hour Worked"].max(),
        line=dict(color="gray", width=2, dash="dash"),
    )
)

# Add annotation for 2015 Base Year
fig.add_annotation(
    x=2015, 
    y=OPH["GDP per Hour Worked"].max(),
    text="2015 = 100 (Base Year)",
    showarrow=False,
    font=dict(size=12, color="gray"),
    xshift=10
)

# Show the chart
fig.show()
```

```python
import plotly.graph_objects as go
import numpy as np

# Fake data
time = np.linspace(0, 10, 100)
n_attributes = 8
angles = np.linspace(0, 2*np.pi, n_attributes, endpoint=False)

fig = go.Figure()

for i, angle in enumerate(angles):
    x = np.cos(angle) * np.ones_like(time)  # radial direction
    y = np.sin(angle) * np.ones_like(time)
    z = time
    values = np.sin(time + i) * 0.5  # some dummy data variation
    fig.add_trace(go.Scatter3d(x=x + values, y=y + values, z=z,
                               mode='lines',
                               name=f'Attribute {i+1}'))

fig.update_layout(scene=dict(
    xaxis_title='Radial X',
    yaxis_title='Radial Y',
    zaxis_title='Time'
))
fig.show()
```
