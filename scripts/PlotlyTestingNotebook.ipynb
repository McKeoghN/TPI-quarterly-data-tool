```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
dataset = pd.read_csv("../out/Dataset.csv")
print(dataset)
```

```python
fig = px.line(dataset, 
              x="Quarter", 
              y=["UK OPH", "Euro Area OPH", "Germany OPH", "Italy OPH"], 
              title="Time Series Comparison")
fig.show()
```

```python
OPH = pd.read_csv("../out/OPH_Processed.csv")

print(OPH)
```

```python
fig = px.line(OPH, 
              x="Year", 
              y="GDP per Hour Worked", 
              color="Country",
              markers=True,
              title="GDP per Hour Worked (2015=100) Over Time")

fig.add_shape(
    go.layout.Shape(
        type="line",
        x0=2015, x1=2015,  # Vertical line at 2015
        y0=OPH["GDP per Hour Worked"].min(), 
        y1=OPH["GDP per Hour Worked"].max(),
        line=dict(color="gray", width=2, dash="dash"),
    )
)

# Add annotation for 2015 Base Year
fig.add_annotation(
    x=2015, 
    y=OPH["GDP per Hour Worked"].max(),
    text="2015 = 100 (Base Year)",
    showarrow=False,
    font=dict(size=12, color="gray"),
    xshift=10
)

# Show the chart
fig.show()
```
